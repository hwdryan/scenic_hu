```python
import csv

# Define the log file name
log_file = "temp"

# Define the output CSV file name
csv_file = "output_data.csv"

# Define IDs of interest
ids_of_interest = {"197", "204"}

# Initialize lists to store extracted data
data = []

# Open the log file and read its contents
with open(log_file, "r+") as file:
    lines = file.readlines()

# 
timestep = None
frame_index = None
# Iterate over each line in the log file
for line in lines:
    line = line.replace("(","")
    line = line.replace(")","")
    line = line.replace(",","")
    line = line.replace(":","")
    line = line.strip().split()
    if all(i in line for i in ['Frame', 'at', 'seconds']):
        timestep = line[3]
        frame_index = line[1]
    # Check if the line contains information for IDs of interest
    elif len(line)>=2 and \
            all(i in line for i in ['Id','Location','Rotation']) and \
            line[1] in ids_of_interest:
        # Extract relevant information
        id_value = line[1]
        location_index = line.index("Location")
        location_value = tuple(map(float, line[location_index+1:location_index+4]))
        rotation_index = line.index("Rotation")
        rotation_value = tuple(map(float, line[location_index+1:location_index+4]))
        
        # Append extracted data to the list
        print([frame_index, id_value, *location_value, *rotation_value])
        data.append([frame_index, timestep, id_value, *location_value, *rotation_value])

# Write extracted data to a CSV file
with open(csv_file, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    # Write header
    writer.writerow(["Frame", "timestep", "Id", "Location_x", "Location_y", "Location_z", "Rotation_x", "Rotation_y", "Rotation_z"])
    # Write data rows
    writer.writerows(data)

print("CSV file has been successfully created!")
```
